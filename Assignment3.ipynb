```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler
```

```python
data= pd.read_csv('bank_dataset.csv')
data
```

```python
labelEncoder = LabelEncoder()
labelEncoder.fit(data["sex"])
data["sex"] = labelEncoder.transform(data["sex"])
labelEncoder.fit(data["region"])
data["region"] = labelEncoder.transform(data["region"])
labelEncoder.fit(data["married"])
labelEncoder.fit(data["car"])
data["car"] = labelEncoder.transform(data["car"])
labelEncoder.fit(data["married"])
data["married"] = labelEncoder.transform(data["married"])
labelEncoder.fit(data["save_act"])
data["save_act"] = labelEncoder.transform(data["save_act"])
labelEncoder.fit(data["current_act"])
data["current_act"] = labelEncoder.transform(data["current_act"])


data
```

```python
dbscan = DBSCAN(eps=1.2 , min_samples=3).fit(data.iloc[:,1:])
dbscan.labels_
```

# Normalizing

```python
normalizer = MinMaxScaler(feature_range=(0,1))
norm_data = normalizer.fit_transform(data.iloc[:,1:])
print(norm_data)
```

```python
dbscan = DBSCAN(eps=1.2 , min_samples=3).fit(norm_data)
print (dbscan.labels_)
```

# Hierarichal Clustering On the unnormalized Data

```python
Agglomerative = AgglomerativeClustering().fit(data.iloc[:,1:])
Agglomerative.labels_
```

# On normalized

```python
Agglomerative = AgglomerativeClustering().fit(norm_data)
print(Agglomerative.labels_)
```
