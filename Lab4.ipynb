```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
weather_data = pd.read_csv("weather_nominal.csv")
weather_data
```

```python
labelEncoder = LabelEncoder()
for i in weather_data.columns:
    if i !='id':
        weather_data[i] = labelEncoder.fit_transform(weather_data[i])
weather_data
```

```python
kmeans = KMeans(n_clusters=2).fit(weather_data.iloc[:,1:])
```

```python
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```
