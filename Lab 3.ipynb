```python
import numpy as np
import pandas as pd
import pyfpgrowth as pf
```

```python
df= pd.read_csv('store_data.csv', header=None)
df
```

```python
recordList=df.values.tolist()
cleanList= [[x for x in sublist if str(x) != 'nan']for sublist in recordList]
cleanList[1]
```

```python
FrequentPattern=pf.find_frequent_patterns(transactions=cleanList,support_threshold=200)
#print(FrequentPattern)
Rules=pf.generate_association_rules(patterns=FrequentPattern,confidence_threshold=0.2)
print(Rules)
```
