```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler
```

```python
columns = ["id","RI","Na","Mg","Al","Si","K","Ca","Ba","Fe","Type"]
data = pd.read_csv("../../Dataset/glass.csv",names = columns,header=0)
```

```python
data
```

```python
labelEncoder = LabelEncoder()
labelEncoder.fit(data["Type"])
data["Type"] = labelEncoder.transform(data["Type"])
```

```python
data
```

```python
dbscan = DBSCAN(eps=2 , min_samples=3).fit(data.iloc[:,1:])
print (dbscan.labels_)
```

```python
Agglomerative = AgglomerativeClustering().fit(data.iloc[:,1:])
print (Agglomerative.labels_)
```
